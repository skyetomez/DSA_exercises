```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dt = 101
n = np.linspace(0,100, dt)
A = 40*n*n
B = 2*n*n*n
df = pd.DataFrame(data ={
                        "A": A,
                        "B": B,
                        "n" : n 
                        }, 
                  index=n,
                  columns=["A","B"]
                  )
df.plot()
plt.axvline(21, color='r')
plt.grid()
plt.xlim([18,50])
plt.title("Plot of A = 40n^2 vs 2n^3")
```

```python
dt = 101
n = np.linspace(0,100, dt)
dA = 80*n
dB = 6*n*n
df2 = pd.DataFrame(data ={
                        "dA": dA,
                        "dB": dB,
                        "n" : n 
                        }, 
                  index=n,
                  columns=["dA","dB"]
                  )
df2.plot()
plt.axvline(21, color='r')
plt.grid()
plt.title("Plot of dA = 80n vs 6n")
```
